## Модель

### Функция для вычисления итоговой sMAPE

```python
def smape(real,pred):
    result=(100*(abs(real-pred)/(abs(real)+abs(pred))/2))
    return result.mean()
```

```python
def ismape(rougher,final):
    return 0.25*rougher+0.75*final 
```

### Подготовка данных

```python
gold_recovery_full[['date','rougher.output.recovery','final.output.recovery']]
```

```python
gold_recovery_train=gold_recovery_train.merge(gold_recovery_full[['date','rougher.output.recovery','final.output.recovery']],how='inner',on='date')
#gold_recovery_train=gold_recovery_train[(gold_recovery_train['rougher.output.recovery'] != 0)&(gold_recovery_train['final.output.recovery'] != 0)]
gold_recovery_train=gold_recovery_train.fillna(method='ffill')
gold_recovery_train.info()
```

```python
target1 = gold_recovery_train['rougher.output.recovery']
target2 = gold_recovery_train['final.output.recovery']
features = gold_recovery_train.drop(['date','rougher.output.recovery','final.output.recovery'], axis=1)
```

```python
features_train, features_valid, target1_train, target1_valid,target2_train, target2_valid = train_test_split(
    features, target1,target2, test_size=0.25,train_size=0.75, random_state=12345)
```

```python
display(features_train.shape,target1_valid.shape,target2_valid.shape)
```

### Стандартизация

```python
scaler = StandardScaler()
scaler.fit(features_train)
features_train_scaled = scaler.transform(features_train)
features_valid_scaled = scaler.transform(features_valid)
```

```python
def ml_train(model):
    
    global model11
    global model22
    
    model11=model
    model22=model
    
    model11.fit(features_train_scaled,target1_train)
    predict1=model11.predict(features_valid_scaled)
    rougher_smape=smape(target1_valid, predict1)
    
    model22.fit(features_train_scaled,target2_train)
    predict2=model22.predict(features_valid_scaled)
    final_smape=smape(target2_valid, predict2)
    
    global all_smape
    
    all_smape=ismape(rougher_smape,final_smape)
    
    smape_scorer=make_scorer(smape)
    
    print('rougher smape =',rougher_smape)
    print('final smape =',final_smape)
    print('Итоговый smape =',all_smape)
    print()
    print('Кросс-валидация =',ismape(cross_val_score(model,features_train_scaled,target1_train,scoring=smape_scorer,cv=5).mean(),cross_val_score(model,features_train_scaled,target2_train,scoring=smape_scorer,cv=5).mean()))
    print()
```

```python
target1_test = gold_recovery_test['rougher.output.recovery']
target2_test = gold_recovery_test['final.output.recovery']
features_test = gold_recovery_test.drop(['date','rougher.output.recovery','final.output.recovery'], axis=1)
```

```python
def ml_test():
    
    predict1=model11.predict(features_test)
    rougher_smape=smape(target1_test, predict1)
    
    predict2=model22.predict(features_test)
    final_smape=smape(target2_test, predict2)
    
    all_smape=ismape(rougher_smape,final_smape)
    
    print('rougher smape =',rougher_smape)
    print('final smape =',final_smape)
    print('Итоговый smape =',all_smape)
```

### Линейная регрессия

```python
model1 = LinearRegression()
ml_train(model1)
```

### Решающее дерево

```python
model2 = DecisionTreeRegressor(random_state=12345)
ml_train(model2)
```

### Cлучайный лес

```python
%%time
best_smape=all_smape
best_model=model2
best_esti=10
for esti in range(10,101,10):
    model3 = RandomForestRegressor(random_state=12345, n_estimators=esti)
    print('Количество деревьев равно ',esti)
    ml_train(model3)
    if all_smape<best_smape:
        best_model=model3
        best_smape=all_smape
        best_esti=esti
print()
print('Лучшее количество деревьев',best_esti)
print('Лучший smape',best_smape)
```

### Dummy

```python
dummy1 = DummyRegressor(strategy='median')
dummy2 = DummyRegressor(strategy='median')
dummy1.fit(features_train_scaled,target1_train)
dummy2.fit(features_train_scaled,target2_train)
dummy_predict1 = dummy1.predict(features_valid_scaled)
dummy_predict2 = dummy2.predict(features_valid_scaled)
dummy_final = 0.25*smape(target1_valid, dummy_predict1) + 0.75* smape(target2_valid, dummy_predict2)
dummy_final
```
